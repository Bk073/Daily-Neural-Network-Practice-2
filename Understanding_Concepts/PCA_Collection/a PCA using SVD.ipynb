```python
# import library
import tensorflow as tf
import numpy as np
import sys, os,cv2
from scipy.misc import imread,imresize
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from imgaug import augmenters as iaa
import imgaug as ia
from scipy.ndimage import zoom
from sklearn.utils import shuffle
import matplotlib.animation as animation
import seaborn as sns

plt.style.use('seaborn-white')
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
```

```python
# try the KPCA on a smaller dimension dataset
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=2000, random_state=123,noise =0.1)
plt.figure(figsize=(8,6))
plt.scatter(X[y==0, 0], X[y==0, 1], color='red',  alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', alpha=0.5)
plt.title('A nonlinear 2Ddataset'); plt.ylabel('y coordinate'); plt.xlabel('x coordinate')
plt.grid(True)
plt.show()
```

```python
# Perform SVD on numpy and projection
U,Sigma,VT = np.linalg.svd(X,full_matrices=False)
Sigma      = np.diag(Sigma)
print(U.shape,Sigma.shape,VT.shape)
print(Sigma[:,0].shape)
X_reduced  = U @ Sigma[:,0]
print(X_reduced.shape)
```

```python
X_reduced = X_reduced[:,None]
plt.scatter(range(len(X_reduced[y==0])),X_reduced[y==0],color='Red')
plt.scatter(range(len(X_reduced[y==1])),X_reduced[y==1],color='Blue')

```
