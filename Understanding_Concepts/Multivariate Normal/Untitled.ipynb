```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mvn
```

```python
cov = np.array([[1,0.0],[0.0,1]])
mu  = np.array([0,2])
print(cov,mu)
```

```python
sampled_distribution  = mvn.rvs(mean = mu,cov = cov,size =1000 )
plt.scatter(sampled_distribution[:,0],sampled_distribution[:,1])
plt.show()
```

```python
cov = np.array([[1,3.0],[0.6,1]])
mu  = np.array([0,2])
print(cov,mu)
sampled_distribution  = mvn.rvs(mean = mu,cov = cov,size =1000 )
plt.scatter(sampled_distribution[:,0],sampled_distribution[:,1])
plt.show()
```
