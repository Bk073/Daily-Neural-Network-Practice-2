```python
# import here
import numpy as np 
import matplotlib.pyplot as plt
np.random.seed(6789)
```

```python
# create layers 
l1 = np.random.randn(100,64) * 0.01
l2 = np.random.randn(64, 32) * 0.01
l3 = np.random.randn(32, 16) * 0.01
l4 = np.random.randn(16, 1)  * 0.01

# https://stackoverflow.com/questions/46411180/implement-relu-derivative-in-python-numpy
def relu(x): return x * (x>0)
def drelu(x):
    x[x<0] = 0
    x[x>=0 ] = 1
    return x

def sigmoid(x): return 1/(1+np.exp(-x))
def dsigmoid(x):return sigmoid(x) * (1-sigmoid(x))

def batchnorm_forward (X, gamma, beta):
    mu     = np.mean(X, axis=0,keepdims=True)
    var    = np.var (X, axis=0,keepdims=True)
    X_norm = (X - mu) / np.sqrt(var + 1e-8)
    out    = gamma * X_norm + beta
    cache  = (X, X_norm, mu, var, gamma, beta)
    return out, cache
def batchnorm_backward(dout, cache):
    X, X_norm, mu, var, gamma, beta = cache

    dgamma = np.sum(dout * X_norm, axis=0)
    dbeta  = np.sum(dout, axis=0)
    
    N, D    = X.shape
    X_mu    = X - mu
    std_inv = 1. / np.sqrt(var + 1e-8)
    dX_norm = dout * gamma
    dvar    = np.sum(dX_norm * X_mu, axis=0) * -.5 * std_inv**3
    dmu     = np.sum(dX_norm * -std_inv, axis=0) + dvar * np.mean(-2. * X_mu, axis=0)
    dX      = (dX_norm * std_inv) + (dvar * 2 * X_mu / N) + (dmu / N)

    return dX, dgamma, dbeta
```

```python
# perform feed forward (relu activation)
x_data = np.random.uniform(0,1,size=(20,100))
random_labels = np.random.binomial(size=(20,1), n=1, p= 0.5)

layer1 = x_data   @ l1; layer1a= relu(layer1)
layer2 = layer1a  @ l2; layer2a= relu(layer2)
layer3 = layer2a  @ l3; layer3a= relu(layer3)
layer4 = layer3a  @ l4; layer4a= sigmoid(layer4)

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(layer1.ravel(),50) ; plt.title('layer ')
plt.subplot(122); plt.hist(layer1a.ravel(),50); plt.title('layer A')
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(layer2.ravel(),50)
plt.subplot(122); plt.hist(layer2a.ravel(),50)
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(layer3.ravel(),50)
plt.subplot(122); plt.hist(layer3a.ravel(),50)
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(layer4.ravel(),50)
plt.subplot(122); plt.hist(layer4a.ravel(),50)
plt.show()
```

```python
# back prop
cost          = 0.5 * (layer4a-random_labels) ** 2
dcost         = (layer4a-random_labels)
dgrad4     = layer3a.T @ (dsigmoid(layer4) * dcost)
dgrad4pass = (dsigmoid(layer4) * dcost)   @ l4.T
dgrad3     = layer2a.T @ (drelu(layer3) * dgrad4pass) 
dgrad3pass = (drelu(layer3) * dgrad4pass) @ l3.T
dgrad2     = layer1a.T @ (drelu(layer2) * dgrad3pass) 
dgrad2pass = (drelu(layer2) * dgrad3pass) @ l2.T
dgrad1     = x_data.T  @ (drelu(layer1) * dgrad2pass) 
dgrad1pass = (drelu(layer1) * dgrad2pass) @ l1.T

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(dgrad4.ravel(),50) ; plt.title('grad ')
plt.subplot(122); plt.hist(dgrad4pass.ravel(),50); plt.title('grad pass')
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(dgrad3.ravel(),50)
plt.subplot(122); plt.hist(dgrad3pass.ravel(),50)
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(dgrad2.ravel(),50)
plt.subplot(122); plt.hist(dgrad2pass.ravel(),50)
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(dgrad1.ravel(),50)
plt.subplot(122); plt.hist(dgrad1pass.ravel(),50)
plt.show()
```

```python
# feed forward with batch norm
layer1 = x_data   @ l1; layer1a= relu(layer1)
layer1a,cache1= batchnorm_forward(layer1a,np.ones(64),np.zeros(64))
layer2 = layer1a  @ l2; layer2a= relu(layer2)
layer2a,cache2= batchnorm_forward(layer2a,np.ones(32),np.zeros(32))
layer3 = layer2a  @ l3; layer3a= relu(layer3)
layer3a,cache3= batchnorm_forward(layer3a,np.ones(16),np.zeros(16))
layer4 = layer3a  @ l4; layer4a= sigmoid(layer4)

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(layer1.ravel(),50) ; plt.title('layer ')
plt.subplot(122); plt.hist(layer1a.ravel(),50); plt.title('layer A')
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(layer2.ravel(),50)
plt.subplot(122); plt.hist(layer2a.ravel(),50)
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(layer3.ravel(),50)
plt.subplot(122); plt.hist(layer3a.ravel(),50)
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(layer4.ravel(),50)
plt.subplot(122); plt.hist(layer4a.ravel(),50)
plt.show()
```

```python
# back prop (bn)
cost          = 0.5 * (layer4a-random_labels) ** 2
dcost         = (layer4a-random_labels)
dgrad4     = layer3a.T @ (dsigmoid(layer4) * dcost)
dgrad4pass = (dsigmoid(layer4) * dcost)   @ l4.T
dgrad4pass,_,_ = batchnorm_backward(dgrad4pass,cache3)

dgrad3     = layer2a.T @ (drelu(layer3) * dgrad4pass) 
dgrad3pass = (drelu(layer3) * dgrad4pass) @ l3.T
dgrad3pass,_,_ = batchnorm_backward(dgrad3pass,cache2)

dgrad2     = layer1a.T @ (drelu(layer2) * dgrad3pass) 
dgrad2pass = (drelu(layer2) * dgrad3pass) @ l2.T
dgrad2pass,_,_ = batchnorm_backward(dgrad2pass,cache1)

dgrad1     = x_data.T  @ (drelu(layer1) * dgrad2pass) 
dgrad1pass = (drelu(layer1) * dgrad2pass) @ l1.T

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(dgrad4.ravel(),50) ; plt.title('grad ')
plt.subplot(122); plt.hist(dgrad4pass.ravel(),50); plt.title('grad pass')
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(dgrad3.ravel(),50)
plt.subplot(122); plt.hist(dgrad3pass.ravel(),50)
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(dgrad2.ravel(),50)
plt.subplot(122); plt.hist(dgrad2pass.ravel(),50)
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(121); plt.hist(dgrad1.ravel(),50)
plt.subplot(122); plt.hist(dgrad1pass.ravel(),50)
plt.show()
```
