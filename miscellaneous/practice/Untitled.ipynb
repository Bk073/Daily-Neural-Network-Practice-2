```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp = np.random.randn(10,64,64)
```

```python
plt.imshow(temp[0])
plt.show
```

```python
print(temp.max((1,2)))
print(temp.min((1,2)))

temp_norm = (temp-temp.min((1,2),keepdims=True))/(temp.max((1,2),keepdims=True)-temp.min((1,2),keepdims=True))
print(temp_norm.max((1,2)))
print(temp_norm.min((1,2)))
```
